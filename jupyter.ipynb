```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the Titanic dataset
df = pd.read_csv(r'C:\Users\PRIYANKA\Downloads\train.csv')

# Data cleaning
# Remove duplicate records
df.drop_duplicates(inplace=True)

# Fill in missing values for age
df['Age'] = df['Age'].fillna(df.groupby(['Pclass', 'Sex'])['Age'].transform(np.median))

# Convert categorical variables to numerical variables
df['Sex'] = df['Sex'].map({'male': 1, 'female': 2})
df['Embarked'] = df['Embarked'].map({'S': 1, 'C': 2, 'Q': 3})

# EDA
# Descriptive statistics
df.describe()

# Visualizations
# Histogram of age
sns.histplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age in Titanic Dataset')
plt.show()

# Bar chart of survival rate by sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex in Titanic Dataset')
plt.show()
```
